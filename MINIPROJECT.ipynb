```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Define the CSV file name
csv_file = 'Spotify Quarterly.csv'

# Error handling for missing file
if not os.path.exists(csv_file):
    print(f"Error: '{csv_file}' not found. Please check the file path.")
    exit()

# Load the data as a DataFrame
df = pd.read_csv(csv_file)

# Inspect first few rows and headers
print("Columns available:", df.columns.tolist())
print(df.head())
```

```python
# Ask the user for the columns to plot
print("Available columns:")
for idx, col in enumerate(df.columns):
    print(f"{idx}: {col}")

selected_indices = input("Enter column numbers you want to plot (comma-separated, e.g. '1,3'): ")
selected_cols = [df.columns[int(idx.strip())] for idx in selected_indices.split(',')]
print("You selected:", selected_cols)
```

```python
# Assume there is a 'Quarter' or 'Date' column to use as x-axis
x_axis = 'Quarter'  # Adjust based on your CSV

for col in selected_cols:
    plt.figure(figsize=(10,6))
    plt.bar(df[x_axis], df[col])
    plt.xlabel(x_axis)
    plt.ylabel(col)
    plt.title(f"{col} per {x_axis}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
